## Missing values

### How to take care of missing data.

### Simple Imputer

```python
# How to take care of missing data
# Simple Imputer method

from sklearn.impute import SimpleImputer 

#assign the function for missing value with replacement
imputer=SimpleImputer(missing_values=np.nan,strategy='mean') # Function is ready

#applicable only numeric data

imputer.fit(x[:,1:3])  # fitting the imputer object

x[:,1:3]=imputer.transform(x[:,1:3]) # replacing the missing values reflect the result of column
```
